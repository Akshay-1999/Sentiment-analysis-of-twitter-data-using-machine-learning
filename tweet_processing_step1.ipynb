```python
#importing the csv file to the program
#reading the data
data = pd.read_csv("/content/drive/MyDrive/ Twitter_data/russia ukraine war 2000.csv")
```

```python
#importing required lib
import pandas as pd
import nltk
import re
import string
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import TweetTokenizer
```

```python
#expanding the dispay of text sms column
pd.set_option('display.max_colwidth', -1)
#using only text column
data= data [['text']]
data.head()
```

```python
#converting the tweets to lower case
data['lower_text'] =  data['text'].str.lower()
data.head()
```

```python
#removing urls from the tweet
data['no_urls'] =data['lower_text'] .apply(lambda x: re.sub(r'https?:\/\/\S+', '', x))
data['no_urls'].apply(lambda x: re.sub(r"www\.[a-z]?\.?(com)+|[a-z]+\.(com)", '', x))
data.head()
```

```python
#removing other linkes from the tweets
data['link_clean'] =data['no_urls'].apply(lambda x: re.sub(r'{link}', '', x))
data['link_clean'].apply(lambda x: re.sub(r"\[video\]", '', x))
data.head()
```

```python
#cleaning other links from the tweet
data['htmlref_clean'] =data['link_clean'].apply(lambda x: re.sub(r'&[a-z]+;', '', x))
data.head()
```

```python
 #removing punctions from the tweet
 data['no_punctions'] =data['htmlref_clean'].apply(lambda x: re.sub(r"[^a-z\s\(\-:\)\\\/\];='#]", '', x))
 data.head()
```

```python
 data['no_hashtag'] =data['no_punctions'].apply(lambda x: re.sub(r'@mention', '', x))
 data['no_hashtag'].apply(lambda x: re.sub(r"www\.[a-z]?\.?(com)+|[a-z]+\.(com)", '', x))
 data.head()
```

```python
#library that contains punctuation
string.punctuation
 #defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
data['clean_punction']= data['no_hashtag'].apply(lambda x:remove_punctuation(x))
data.head()
```

```python
#defining function for tokenization
tknzr = TweetTokenizer()
data['tokens'] = data['clean_punction'].apply(tknzr.tokenize)
data.head()
```

```python
#expanding the dispay of tokens  column
pd.set_option('display.max_colwidth', 1)
#using only tokens column
data= data [['tokens']]
data.head()
```

```python
 #Stop words present in the library
 nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
stopwords[:]
print(stopwords)
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
#applying the function
data['no_stopwords']= data['tokens'].apply(lambda x:remove_stopwords(x))
data.head()
```

```python
#importing the Stemming function from nltk library
#defining the object for stemming
porter_stemmer = PorterStemmer()
```

```python
#importing the Stemming function from nltk library
#defining the object for stemming
porter_stemmer = PorterStemmer()
#defining a function for stemming
def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text
data['msg_stemmed']=data['no_stopwords'].apply(lambda x: stemming(x))
data.head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
#defining the function for lemmatization
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
data['msg_lemmatized']=data['no_stopwords'].apply(lambda x:lemmatizer(x))
data.head()
```

```python
#expanding the dispay of msg_lemmatized  column
pd.set_option('display.max_colwidth', 1)
data= data [['msg_lemmatized']] 
data.head()
```

```python

# print in tabular format
result = data.to_string(index = False)
print(result)
```
